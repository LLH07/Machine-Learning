What the flying f is p-value?  
A p-value is the probability that random chance generated the data, or sth else that is equal or rarer.  
這句話可以拆成三部分 (也就是 p-value 組成的三部分)，分別為:  
- the probability that random chance generated the data  
- or sth else that is equal  
- or sth else that is rarer  
請看下方圖例:

example of p-value

We are now finding the <font color='red'> p-value </font> for ```HH```:  
- the probability that random chance (隨機) generated the data (產生的目標資料 [ie ```HH```]) = 所有可能中產生 ```HH``` 的機率 = $\frac{1}{4} = 0.25$    
- or sth else that is equal = 跟 ```HH``` 有一樣機率發生的情況 的機率 = ```TT``` 的機率 = $\frac{1}{4} = 0.25$
- or sth else that is rarer = 在此例為 0，因為 $\frac{1}{4} = 0.25$ 已經是最小的機率了，其他如 ```HT``` 或 ```TH``` 的機率均為 50%
---
因此 p-value for ```HH``` = 0.25+0.25+0=0.5。  
同時，the probability of getting ```HH``` = 0.25 $\neq$ the p-value 0f getting ```HH``` = 0.50

上面的例子考慮的是 discrete case，那如果是 continuous case 呢?  
可以藉由 density 可以藉由 ```density``` (即下圖) 取得資訊。    
The area under the curve indicates the probability that a person will have a height within a range of possible values. 舉例來說，一個人有 95% 的機率 (the probability)，他的身高 (a person have a height) 落在 142-169 公分 (within a range of possible values). 
--- 
現在假設我想計算 <font color='red'> the p-value for someone who is 142 cm </font>  :    
$p-value = 0.025 + 0.025 = 0.05$。其中第一個 0.025 是身高小於 142 cm 的機率，第二個 0.025 是身高大於 169 cm 的機率。  
Case 2: <font color='red'> the p-value for someone who is 155.4-156 cm </font>  :    
藉由 ```density```，身高介於 155.4-156 cm 的機率是 4%，進一步代表 the probability that is equal or rarer 是 0.48 (身高小於 155.4 cm)+ 0.48 (身高大於 156 cm)= 0.96。因此，the p-value for someone who is 155.4-156 cm = 0.04 + 0.96 = 1 !!
